```python
def get_all_pages_structure(pages):    
    
    all_pages = []
    
    headline1 = None
    headline2 = None

    terminating_symbols = ('.', ':', '!', '?', ';')
    
    last_span = 0   # flag for what the last span was (0 = paragraph, 1 = h1, 2 = h2)
    
    page_structure = []
    
    for idx, page in enumerate(pages):
        print(f"Extracting {idx} page!")
        text = page.get_text('dict')
        
        # page_structure = []

        for block in text['blocks']:
            if 'lines' in block:
                for line in block['lines']:

                    for span in line['spans']:
                        span_text = span['text']
                        span_font = span['font']
                        span_size = span['size']
                        span_dic = {}

                        
                        if span_text.isnumeric():
                            continue
                        
                        if span_font == font_header and span_size == size_header1:
                            last_span = 1
                            headline1 = span_text
                            headline2 = None
                            continue
                        
                        if span_font == font_header and span_size == size_header2:
                            last_span = 2
                            headline2 = span_text
                            continue
                        
                        span_dic['paragraph'] = span_text
                        
                        if page_structure and not page_structure[-1]['paragraph'].rstrip().endswith(terminating_symbols) and last_span == 0:
                            page_structure[-1]['paragraph'] = page_structure[-1]['paragraph'] + span_text
                            continue
                            
                        span_dic['header1'] = headline1
                        span_dic['header2'] = headline2
                        
                        last_span = 0
                        
                        page_structure.append(span_dic)
    
        all_pages.append(page_structure)
    
    return page_structure
```
